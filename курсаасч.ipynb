# Описание столбцов

Описания столбцов:

Школа: название посещаемой школы (например, MS).

Пол: пол учащегося (например, M — мужчина, F — женщина).

Возраст: возраст учащегося.

Адрес: Тип проживания (U для городского, R для сельского).

Family_Size: размер семьи (GT3 — больше 3, LE3 — меньше или равно 3).

Родительский_статус: условия проживания родителей (A — совместное проживание, T — раздельное проживание).

Mother_Education: уровень образования матери (от 0 до 4).

Образование_отца: уровень образования отца (от 0 до 4).

Mother_Job: Тип работы, которую выполняет мать.

Father_Job: Тип работы, которую выполняет отец.

Причина_выбора_школы: причина выбора школы (например, курс).

Опекун: Опекун учащегося (например, мать).

Время в пути: время, затрачиваемое на дорогу до школы (в минутах).

Study_Time: Еженедельные учебные часы (с 1 по 4).

Number_of_Failures: количество прошлых неудач в классе.

Школьная_поддержка: получает ли учащийся дополнительную образовательную поддержку (да/нет).

Family_Support: Семья оказывала образовательную поддержку (да / нет).

Extra_Paid_Class: участие в дополнительных платных занятиях (да/нет).

Внеклассные_занятия: участие во внеклассных мероприятиях (да/нет).

Посещал_детский_сад: посещение детского сада (да/нет).

Желание_получить_высшее_образование: желание получить высшее образование (да/нет).

Доступ к интернету: наличие интернета дома (да / нет).

In_Relationship: статус романтических отношений (да / нет).

Семейные_отношения: качество семейных отношений (по шкале от 1 до 5).

Свободное_время: количество свободного времени после школы (по шкале от 1 до 5).

Выход_в_свет: частота выходов в свет с друзьями (по шкале от 1 до 5).

Потребление алкоголя в выходные: потребление алкоголя в выходные (по шкале от 1 до 5).

Потребление алкоголя в будние дни: потребление алкоголя в будние дни (по шкале от 1 до 5).

Состояние здоровья: оценка здоровья учащегося (по шкале от 1 до 5).

Количество_пропусков: общее количество пропусков занятий в школе.

Grade_1: оценка, полученная при первом оценивании.

Grade_2: оценка, полученная при втором оценивании.

Final_Grade: получена итоговая оценка (G3).

Выбыл_из_школы: индикатор того, выбыл ли ученик из школы (истина/ложь).

# Генерация синтетических данных


```python

# Определяем целевую длину
target_length = 50000
current_length = len(df)

# Создание синтетических данных
synthetic_data = pd.DataFrame()

for _ in range(target_length // current_length):
    # Создаем новый DataFrame, добавляя шум к существующим данным
    new_data = df.copy()
    new_data['Age'] += np.random.randint(-1, 2, size=current_length)  
    new_data['Free_Time'] += np.random.randint(-1, 2, size=current_length)
    new_data['Family_Relationship'] += np.random.randint(-1, 2, size=current_length)
    new_data['Final_Grade'] += np.random.randint(-2, 3, size=current_length)  
    new_data['Number_of_Absences'] += np.random.randint(-1, 2, size=current_length)
    new_data['Study_Time'] += np.random.randint(-1, 2, size=current_length)


    # Убедимся, что числовые значения остаются в разумных пределах
    new_data['Age'] = np.clip(new_data['Age'], 15, 22)
    new_data['Free_Time'] = np.clip(new_data['Free_Time'], 1, 5)
    new_data['Family_Relationship'] = np.clip(new_data['Family_Relationship'], 1, 5)
    new_data['Final_Grade'] = np.clip(new_data['Final_Grade'], 0, 19)
    new_data['Number_of_Absences'] = np.clip(new_data['Number_of_Absences'], 0, 32)
    new_data['Study_Time'] = np.clip(new_data['Study_Time'], 1, 4)


    # Добавляем шум к категориальным переменным (с некоторой вероятностью меняем их)
    for col in ['Wants_Higher_Education', 'Family_Size','Extra_Curricular_Activities','Extra_Paid_Class']:
        flip = np.random.rand(current_length) < 0.2  # 20% изменения
        new_data.loc[flip, col] = np.random.choice(df[col].unique(), size=flip.sum())

    synthetic_data = pd.concat([synthetic_data, new_data], ignore_index=True)

 
df= synthetic_data.head(target_length)

df.head()

```

```python


# Создание новой колонки для категорий
def categorize_relationship(value):
    if value in [1, 2]:
        return 'Плохо'
    elif value == 3:
        return 'Средне'
    elif value in [4, 5]:
        return 'Отлично'
    else:
        return 'Неизвестно'

df['Relationship_Category'] = df['Family_Relationship'].apply(categorize_relationship)

# Создание сводной таблицы для каждой категории
cross_tab_poor = pd.crosstab(df[df['Relationship_Category'] == 'Плохо']['Final_Grade'], 
                              df['Relationship_Category'])

cross_tab_average = pd.crosstab(df[df['Relationship_Category'] == 'Средне']['Final_Grade'], 
                                 df['Relationship_Category'])

cross_tab_excellent = pd.crosstab(df[df['Relationship_Category'] == 'Отлично']['Final_Grade'], 
                                   df['Relationship_Category'])


fig, axes = plt.subplots(3, 1, figsize=(20, 5))

# Визуализация для "Плохо"
sns.heatmap(cross_tab_poor.T, annot=True, cmap='Blues', fmt='d', ax=axes[0], cbar=False)
axes[0].set_title('Отношения в семье: Плохо')



# Визуализация для "Средне"
sns.heatmap(cross_tab_average.T, annot=True, cmap='Blues', fmt='d', ax=axes[1], cbar=False)
axes[1].set_title('Отношения в семье: Средне')


# Визуализация для "Отлично"
sns.heatmap(cross_tab_excellent.T, annot=True, cmap='Blues', fmt='d', ax=axes[2], cbar=False)
axes[2].set_title('Отношения в семье: Отлично')



plt.tight_layout()
plt.show()

```

```python

def categorize_relationship(value):
    if value ==1:
        return 'Мало'
    elif value in [2,3]:
        return 'Средне'
    elif value ==4:
        return 'Много'
    else:
        return 'Неизвестно'

df['Study_time_Category'] = df['Study_Time'].apply(categorize_relationship)


pivot_table = df.pivot_table(index='Study_time_Category', columns='Relationship_Category', aggfunc='size', fill_value=0)

# Устанавливаем порядок столбцов
desired_order = ['Плохо', 'Средне', 'Отлично']
pivot_table = pivot_table.reindex(columns=desired_order)

desired_index_order = ['Много','Средне','Мало'] 
pivot_table = pivot_table.reindex(index=desired_index_order)
# Теперь pivot_table будет иметь столбцы в нужном порядке
print(pivot_table)

# Визуализация
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt='d', cmap='Blues')
plt.title('Study Time vs Relationship Category')
plt.xlabel('Relationship Category')
plt.ylabel('Study Time Category')
plt.show()    
pivot_table    
```
