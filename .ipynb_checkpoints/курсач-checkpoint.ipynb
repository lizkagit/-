# Чтение файла и работа с датасетом


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pandas.plotting import scatter_matrix
df=pd.read_csv('данные о кредитах.csv')
df.columns = df.columns.str.strip()

df
```

```python
df.info()

```

## Распределение кредитного рейтинга



```python
custom=['#6ACC64','#D65F5F']
sns.set(style = "darkgrid")
plt.figure(figsize=(12,6))
sns.histplot(data=df, x='cibil_score', hue='loan_status', multiple='stack', bins=20,palette=custom)
plt.title('Распределение кредитного рейтинга по статусу кредита ')
plt.xlabel('CIBIL Score')
plt.ylim(0,250)
plt.legend(['Отказано','Одобрено'])
plt.ylabel('Число кредитов')
plt.show()
```

## Срок кредита

```python
df_1 = df.groupby('loan_term')['loan_status'].value_counts().unstack().fillna(0)
plt.figure(figsize=(10,5)) 
df_1.plot(kind='bar', stacked=False,color=['green','red'])
plt.ylabel('Кол-во кредитов')
plt.xlabel('Срок кредита в годах')
plt.title('Количество одобренных и неодобренных кредитов в зависимости от срока кредита')
plt.legend(['Одобрено','Отказано'])
plt.show()
df_1
```

## Сумма кредита


```python
plt.figure(figsize=(12,6))
sns.histplot(data=df, x='loan_amount', hue='loan_status', multiple='stack', bins=20,palette=custom)
 
plt.xlabel('Сумма кредита')
plt.title('Распределение суммы кредита по статусу кредита')
plt.legend(['Отказано','Одобрено'])
plt.ylabel('Число одобренных кредитов')
plt.show()


 
```

##  Образование

```python
education_status = df.groupby('education')['loan_status'].value_counts(normalize=True).unstack().fillna(0)
education_status = education_status * 100

plt.figure(figsize=(12, 6))
education_status.plot(kind='bar', stacked=False,color=['green','red'])
plt.title('Одобрение кредитов по уровню образования')
plt.xlabel('Уровень образования')
plt.ylabel('Процент одобренных кредитов')
plt.legend(title='Статус кредита', labels=['Одобрен','Не одобрен'])
plt.show()
education_status
```

## Статус занятости

```python
employment_status = df['self_employed'].value_counts(normalize=True) * 100
plt.figure(figsize=(6, 4))
employment_status.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Процент одобренных кредитов по статусу занятости')
plt.ylabel('')
plt.show()
```

##  Годовой доход заемщика

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='loan_status', y='income_annum', data=df)
plt.title('Годовой доход заемщиков по статусу кредита')
plt.xlabel('Статус кредита')
plt.ylabel('Годовой доход')
plt.show()
```

## Количество иждивенцев

```python
education_status = df.groupby('no_of_dependents')['loan_status'].value_counts(normalize=True).unstack().fillna(0)
education_status = education_status * 100

plt.figure(figsize=(12, 6))
education_status.plot(kind='bar', stacked=False,color=['green','red'])
plt.title('Одобрение кредитов по кол-ву иждивенцев')
plt.xlabel('Кол-во иждивенцев')
plt.ylabel('Процент одобренных кредитов')
plt.legend(title='Статус кредита', labels=['Одобрен','Не одобрен'])
plt.show()
```

# Стоимость активов

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='residential_assets_value', hue='loan_status', multiple='stack', bins=15,palette=custom)
plt.title('Распределение кредитного рейтинга по стоимости жилых активов')
plt.xlabel('стоимость жилых активов')
plt.legend(['Отказано','Одобрено'])
plt.ylabel('Число кредитов')
 
plt.show()

plt.figure(figsize=(12,6))
sns.histplot(data=df, x='commercial_assets_value', hue='loan_status', multiple='stack', bins=15,palette=custom)
plt.title('Распределение кредитного рейтинга по стоимости коммерческих активов')
plt.xlabel('стоимость коммерческих активов')
plt.legend(['Отказано','Одобрено'])
plt.ylabel('Число кредитов')
 
plt.show()

plt.figure(figsize=(12,6))
sns.histplot(data=df, x='luxury_assets_value', hue='loan_status', multiple='stack', bins=15,palette=custom)
plt.title('Распределение кредитного рейтинга по стоимости предметов роскоши')
plt.xlabel('стоимость предметов роскоши')
plt.legend(['Отказано','Одобрено'])
plt.ylabel('Число кредитов')
 
plt.show()

plt.figure(figsize=(12,6))
sns.histplot(data=df, x='bank_asset_value', hue='loan_status', multiple='stack', bins=15,palette=custom)
plt.title('Распределение кредитного рейтинга по стоимости банковских активов')
plt.xlabel('стоимость банковских активов')
plt.legend(['Отказано','Одобрено'])
plt.ylabel('Число кредитов')
 
plt.show()
```

# Кореляционная матрица

```python
 df_1=df[['cibil_score','income_annum','loan_amount',
         'loan_term','no_of_dependents','bank_asset_value',
         'luxury_assets_value','residential_assets_value','commercial_assets_value']]

 
correlation_matrix = df_1.corr()
 
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Корреляционная матрица')
plt.show()

```
